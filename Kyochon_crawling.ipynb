```python
from bs4 import BeautifulSoup
import urllib.request
from selenium import webdriver
```

```python
from selenium.webdriver.common.by import By
import time
```

```python
result = []
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(10)
kyochonUrl = 'http://www.kyochon.com/shop/domestic.asp'
driver.get(kyochonUrl)

# sido_list = driver.find_elements(By.CSS_SELECTOR, "#sido1 > option")# 시도 검색 press
# length = len(sido_list) # 시/도 갯수
# print(length)
# time.sleep(0.5)

sido1 = 1
while(sido1 < 18):
    time.sleep(0.5)
    sido_list = driver.find_elements(By.CSS_SELECTOR, "#sido1 > option")
    sido_list[sido1].click()
    time.sleep(0.5)

    # 시/군/구 검색
    sido2 =  driver.find_element(By.CSS_SELECTOR, "#sido2")
    sido2.click()
    time.sleep(0.5)

    sido2_list = driver.find_elements(By.CSS_SELECTOR, "#sido2 > option")
    length2 = len(sido2_list) # 시군구 갯수
#     print(length2)

    for sido2 in range(1, length2 + 1):
        kyochon_url = 'http://www.kyochon.com/shop/domestic.asp?sido1=%d&sido2=%d&txtsearch=' %(sido1,sido2)
        print(kyochon_url)
        html = urllib.request.urlopen(kyochon_url)
        soupKyochon = BeautifulSoup(html, 'html.parser')

        tag_tbody = soupKyochon.find("div", "shopSchList")
        for llist in tag_tbody.select("ul > li"):
            if llist.get_text() != '검색결과가 없습니다.':
                store = llist.select_one("a > span > strong").string
                em = llist.select_one("a > span > em").get_text()
                store_address = em.strip().splitlines()[0]
                sido = store_address.split(' ')[0]
                gungu = store_address.split(' ')[1]
                result.append([store]+[sido]+[gungu]+[store_address])
                print(store)
                print(store_address)
                print(sido)
                print(gungu)
                print('--------------------')
        print('================')
        
    sido1 += 1
```

```python
import pandas as pd
```

```python
kyochon_tbl = pd.DataFrame(result, columns = ('store', 'sido', 'gungu', 'store_address'))
kyochon_tbl.to_csv('kyochon.csv', encoding = 'cp949', mode = 'w', index = True)
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(10)
kyochonUrl = 'http://www.kyochon.com/shop/domestic.asp'
driver.get(kyochonUrl)

# sido1 = driver.find_element(By.CSS_SELECTOR, "#sido1")
# sido1.click()

sido_list = driver.find_elements(By.CSS_SELECTOR, "#sido1 > option")# 시도 검색 press
length = len(sido_list) # 시/도 갯수
time.sleep(0.5)

# print(length)

for i in range(1, length):
    time.sleep(0.5)
    sido_list = driver.find_elements(By.CSS_SELECTOR, "#sido1 > option")
    sido_list[i].click()
    time.sleep(0.5)

    # 시/군/구 검색
    sido2 =  driver.find_element(By.CSS_SELECTOR, "#sido2")
    sido2.click()
    time.sleep(0.5)

    sido2_list = driver.find_elements(By.CSS_SELECTOR, "#sido2 > option")
    length2 = len(sido_list) # 시군구 갯수

    for j in range(1, length2):
        sido2 =  driver.find_element(By.CSS_SELECTOR, "#sido2")
        sido2.click()
        time.sleep(0.5)

        sido2_list = driver.find_elements(By.CSS_SELECTOR, "#sido2 > option")
        sido2_list[j].click()
        time.sleep(0.5)
        driver.execute_script('search()')  # js 실행
        time.sleep(0.5)
        
        html = urllib.request.urlopen(kyochonUrl)
        soupKyochon = BeautifulSoup(html, 'html.parser')
        tag_tbody = soupKyochon.find("div", "shopSchList")
        
        for llist in tag_tbody.select("ul > li"):
            store = llist.select_one("a > span > strong").string
            em = llist.select_one("a > span > em").get_text()
            store_address = em.strip().splitlines()[0]
            sido = store_address.split(' ')[0]
            gungu = store_address.split(' ')[1]
            print(store)
            print(store_address)
            print(sido)
            print(gungu)
            print('--------------------')
            time.sleep(0.5)
        time.sleep(0.5)
        print('======================')

```
